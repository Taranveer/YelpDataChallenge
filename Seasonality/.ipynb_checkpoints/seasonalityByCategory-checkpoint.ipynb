```python
def load(filename):
    file = open(filename,'rb')
    obj = pickle.load(file)
    file.close()
    return obj

def dump(obj,filename):
    filehandler = open(filename,"wb")
    pickle.dump(obj,filehandler)
    filehandler.close()
```

# Time Series Exploration of Ratings (Individual Restaurant)

```python
# plot stars against date for business_id = 0Wy4gw8krao9nGq-sHWFSQ
biz = pit.loc[pit['business_id'] == '0Wy4gw8krao9nGq-sHWFSQ']
# process date
#dates = biz['date']
#biz['date'] = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in dates]
# take average daily ratings
#biz_ave = averageStars(biz, 'date')
```

```python
uniqueDim = list(set(biz['date']))
#sort date
uniqueDim = sorted(uniqueDim)
aveRatings = []
for d in uniqueDim:
    totalRatings = sum(biz.loc[biz['date'] == d]['stars_x'])
    length = float(len(biz.loc[biz['date'] == d])) 
    ave = totalRatings/length
    aveRatings.append(ave)
```

```python
uniqueDim = list(set(biz['date']))
#sort date
uniqueDim = sorted(uniqueDim)
```

```python
def averageStars(df, dimension):
    uniqueDim = sorted(list(set(df[dimension])))
    biz_s = [(sum(df.loc[df[dimension] == d]['stars_x']))/float(len(df.loc[df[dimension] == d])) for d in uniqueDim]
    return biz_s
```

```python
aveRatings = averageStars(biz, 'date')
```

```python
plt.plot(uniqueDim, aveRatings)
plt.show()
```

# Time Series Exploration of Ratings (By Restaurant Category)

```python
# aggregate average stars for each category against date
categories = (pit['categories'])
```

```python
aggregateCategories = []

for i in range(0, categories.size):
    parts = categories.iloc[i]
    for j in parts:
        aggregateCategories.append(j)

uniqueCategories = list(set(aggregateCategories))
```

```python
len(uniqueCategories)
```

```python
# get frequency of each category
freqCategories = []
for d in uniqueCategories:
    count = aggregateCategories.count(d)
    freqCategories.append([count, d])
```

```python
freqCategories = sorted(freqCategories, reverse = True)
# Top 20 Most Frequent categories
freqCategories[:20]
```

```python
# store Top 20 categories
top20cate = []
for c in freqCategories[:20]:
    top20cate.append(c[1])
```

```python
# for each cateogry, calculate daily average ratings
df_Pubs = pd.DataFrame(columns=list(pit))
d = 'Pubs'

for i in range(0, len(pit)):
    if (d in pit['categories'].iloc[i]):
        # append to dataframe
        df_Pubs = df_Pubs.append(pit.iloc[i])

avePubsRatings = averageStars(df_Pubs, 'date')
datesPubs = sorted(list(set(df_Pubs['date'])))
```

```python
# for each cateogry, calculate daily average ratings
df_Chinese = pd.DataFrame(columns=list(pit))
d = 'Chinese'

for i in range(0, len(pit)):
    if (d in pit['categories'].iloc[i]):
        # append to dataframe
        df_Chinese = df_Chinese.append(pit.iloc[i])

aveChineseRatings = averageStars(df_Chinese, 'date')
datesChinese = sorted(list(set(df_Chinese['date'])))
```

```python
axes = plt.gca()
axes.set_ylim([0,5])
plt.plot(datesPubs, avePubsRatings)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 40
fig_size[1] = 7
plt.rcParams["figure.figsize"] = fig_size
plt.figtext(.5,.9,"Daily Average Ratings for Pubs in Pittsburgh", fontsize=50, ha='center')
plt.show()
```
